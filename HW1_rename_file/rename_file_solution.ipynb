```python
import time
import os
import shutil
from datetime import datetime
```

```python
path = './files'
if os.path.exists(path):
    shutil.rmtree(path)
    os.mkdir(path)
for t in range(20):
    with open (f'{path}/test_{datetime.now().strftime("%Y%m%dT%H%M%S")}.txt', "w") as f:
        f.write(f'{datetime.now()}')
    time.sleep(1)
```
